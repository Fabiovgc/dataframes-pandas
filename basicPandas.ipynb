# Pandas Basico

### Importando Pandas

```python
import pandas as pd
```

### Criando um Dataframe a partir de um dicionario 
##### (Tabela dentro do Python)

```python
# dataframe = pd.DataFrame()
venda = {'data':['15/02/2024', '16/02/2024'],
         'valor': ['500', '300'],
         'produto': ['produto_1','produto_2'],
         'qtde':['50','70'],
}

tabela_vendas = pd.DataFrame(venda)
```

### Visualização de Dados
##### - `print`
##### - `display` (No Jupyter o `display` dispões de uma interação com cores quando o mouse está por cima e o head da atbela em negrito)


```python
print(tabela_vendas)
```

### Importando arquivos e bases de dados

```python
tabela_vendas = pd.read_excel("Vendas.xlsx")
display(tabela_vendas)
```

### Resumo de visualização de Dados
##### - `head` (Mostra o inicio da tabela e podemos definir quantas linhas exibidas como parametro de `head`)
##### - `shape`
##### - `describe`

```python
display(tabela_vendas.head(10))
display(tabela_vendas.shape)
display(tabela_vendas.describe())
```

### Pegar 1 coluna (pd.series) - Uma serie do Pandas é uma COLUNA
##### 1 Serie = ['`nome da coluna`'] / 2 Series = [['`nome da coluna`']]

```python
produtos = tabela_vendas[['Produto', 'ID Loja', 'Data']]
display(produtos)
```

### Filtrar por linhas

```python
#Pegar linha 1
print('Apresentando dados da linha 1')
display(tabela_vendas.loc[1])
"\n"
"\n"
print('Apresentando dados da linha 1 até linha 5')
#Pegar da linha 1 a 5 -> Linha que corresponde a uma condição
display(tabela_vendas.loc[1:5])
"\n"
"\n"
print('Apresentando dados de todas as linhas do Norte Shopping')
# Todas as linhas onde a Loja é Norte Shopping
# tabela_vendas['ID Loja'] == 'Norte Shopping' -> Essa linha passamos dentro do parametro do `loc`
vendas_norteshopping_df = tabela_vendas.loc[tabela_vendas['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_df)
```

### Pegar linhas e colunas específicas usando o Metodo `loc`

```python
vendas_norteshopping_df = tabela_vendas.loc[tabela_vendas['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_df)
```

### Procurando um valor específico

```python
display(tabela_vendas.loc[1, ['Produto']])
```

### Adicionar uma nova Coluna
##### - A partir de uma coluna que já existe
##### - Criar coluna com valor padrão

```python
# Que ja existe
tabela_vendas['Comissao'] = tabela_vendas['Valor Final'] * 0.05

# Criar uma coluna com valor padrão
tabela_vendas.loc[:, "Imposto"] = 0
display(tabela_vendas)
```

### Adicionando Linha
##### - Linhas de um complemento de base de dados

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
#display(vendas_dez_df)

tabela_vendas = pd.concat([tabela_vendas, vendas_dez_df], ignore_index=True)
display(tabela_vendas)
```

### Excluir linhas e colunas
##### - Deletar linhas/colunas vazias
##### - Deletar linhas que possuem valores vazios
##### - Preencher valores vazios (média e ultimo valor)

```python
# Excluindo coluna Imposto
print('Excluindo coluna IMPOSTO')
tabela_vendas = tabela_vendas.drop("Imposto", axis=1)
display(tabela_vendas)
"\n"
"\n"

#Excluindo linha 0
print('Excluindo LINHA 0')
tabela_vendas = tabela_vendas.drop(0, axis=0)
display(tabela_vendas)
```

### Tratando valores Vazios ( NaN )
##### - Ddeletar linhas/colunas vazias
##### - Ddeletar linhas que possuem valores vazios
##### - Preencher valores vazios (média e ultimo valor)

```python
# Deletando linhas e colunas completamente vazias
#tabela_vendas = tabela_vendas.dropna(how='all', axis=1)

# Deletar linhas que possuem pelo menos 1 valor vazio
#tabela_vendas = tabela_vendas.dropna()

# Preencher valores vazios

# Preencher com media da coluna
tabela_vendas['Comissao'] = tabela_vendas['Comissao'].fillna(tabela_vendas['Comissao'].mean())
display(tabela_vendas)

# Preencher com o ultimo valor
tabela_vendas = tabela_vendas.ffill()
display(tabela_vendas)
```

### Calcular indicadores
##### - Group By
##### - Value Counts - quantas `linhas` eu tenho por `coluna`

```python
# VALUE COUNTS = Quantas transações eu tive por loja ? (QUANTAS LINHAS EU TENHO POR COLUNA)
transacoes_por_loja = tabela_vendas['ID Loja'].value_counts()
display(transacoes_por_loja)

# GROUP BY = Agrupar informações e somar o valor final
faturamento_produto = tabela_vendas[['Produto', 'Valor Final']].groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)

# GROUP BY = Agrupar informações e achar o valor medio
faturamento_produto = tabela_vendas[['Produto', 'Valor Final']].groupby('Produto').mean()
display(faturamento_produto)
```

### Mesclar 2 `DataFrames`
##### - Procurar informações de um `DataFrame` em outro

```python
# Criando o DataFrame para a tabela de Gerentes
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

### Mesclando o `DataFrame` (Tabela) de Gerentes com o `DataFrame` (Tabela) de Vendas

```python
tabela_vendas = tabela_vendas.merge(gerentes_df)
display(tabela_vendas)
```
